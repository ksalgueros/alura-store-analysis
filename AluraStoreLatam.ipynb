### Importaci√≥n de datos



````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
````

#1. An√°lisis de facturaci√≥n



````python
# Facturaci√≥n por tienda
facturacion1 = tienda['Precio'].sum()
facturacion2 = tienda2['Precio'].sum()
facturacion3 = tienda3['Precio'].sum()
facturacion4 = tienda4['Precio'].sum()

# Mostrar resultados
print("üí∞ Facturaci√≥n total por tienda:")
print(f"Tienda 1: ${facturacion1:,.2f}")
print(f"Tienda 2: ${facturacion2:,.2f}")
print(f"Tienda 3: ${facturacion3:,.2f}")
print(f"Tienda 4: ${facturacion4:,.2f}")
````

````python
# Aseg√∫rate de que la columna de fecha est√© en formato datetime
tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True)

# Extraer el nombre del mes (en espa√±ol si tu entorno lo permite)
tienda['Mes'] = tienda['Fecha de Compra'].dt.month_name()

# Agrupar por nombre del mes y sumar la facturaci√≥n
facturacion_mensual = tienda.groupby('Mes')['Precio'].sum().sort_values(ascending=False)

# Mostrar resultados
print("üìÖ Facturaci√≥n mensual por nombre de mes:")
print(facturacion_mensual)
````

# 2. Ventas por categor√≠a

````python
# Agrupar por categor√≠a de producto y sumar los precios
facturacion_por_categoria = tienda.groupby('Categor√≠a del Producto')['Precio'].sum().sort_values(ascending=False)

# Mostrar resultados
print("üì¶ Facturaci√≥n por categor√≠a de producto:")
print(facturacion_por_categoria)
````

# 3. Calificaci√≥n promedio de la tienda

1.   Elemento de lista
2.   Elemento de lista




````python
# Calcular la calificaci√≥n promedio de cada tienda
calificacion1 = tienda['Calificaci√≥n'].mean()
calificacion2 = tienda2['Calificaci√≥n'].mean()
calificacion3 = tienda3['Calificaci√≥n'].mean()
calificacion4 = tienda4['Calificaci√≥n'].mean()

# Mostrar resultados
print("‚≠ê Calificaci√≥n promedio por tienda:")
print(f"Tienda 1: {calificacion1:.2f}")
print(f"Tienda 2: {calificacion2:.2f}")
print(f"Tienda 3: {calificacion3:.2f}")
print(f"Tienda 4: {calificacion4:.2f}")
````

### #3 - Productos m√°s y menos vendidos

````python
# Contar cu√°ntas veces se vendi√≥ cada producto
ventas_por_producto1 = tienda['Producto'].value_counts()
ventas_por_producto2 = tienda2['Producto'].value_counts()
ventas_por_producto3 = tienda3['Producto'].value_counts()
ventas_por_producto4 = tienda4['Producto'].value_counts()

# Mostrar los m√°s vendidos
print("üìà Productos m√°s vendidos por tienda:")
print("Tienda 1:", ventas_por_producto1.head(3))
print("Tienda 2:", ventas_por_producto2.head(3))
print("Tienda 3:", ventas_por_producto3.head(3))
print("Tienda 4:", ventas_por_producto4.head(3))

# Mostrar los menos vendidos
print("\nüìâ Productos menos vendidos por tienda:")
print("Tienda 1:", ventas_por_producto1.tail(3))
print("Tienda 2:", ventas_por_producto2.tail(3))
print("Tienda 3:", ventas_por_producto3.tail(3))
print("Tienda 4:", ventas_por_producto4.tail(3))
````

# 5. Env√≠o promedio por tienda

````python
# Calcular el env√≠o promedio por tienda
envio1 = tienda['Costo de env√≠o'].mean()
envio2 = tienda2['Costo de env√≠o'].mean()
envio3 = tienda3['Costo de env√≠o'].mean()
envio4 = tienda4['Costo de env√≠o'].mean()

# Mostrar resultados
print("üöö Env√≠o promedio por tienda:")
print(f"Tienda 1: ${envio1:.2f}")
print(f"Tienda 2: ${envio2:.2f}")
print(f"Tienda 3: ${envio3:.2f}")
print(f"Tienda 4: ${envio4:.2f}")
````

# 6. Facturaci√≥n por m√©todo de pago


````python
# Load the data if it's not already loaded
try:
    tienda.head()
except NameError:
    import pandas as pd
    url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
    tienda = pd.read_csv(url)


# Agrupar por m√©todo de pago y sumar la facturaci√≥n
facturacion_por_pago = tienda.groupby('M√©todo de pago')['Precio'].sum().sort_values(ascending=False)

# Mostrar resultados
print("üí≥ Facturaci√≥n por m√©todo de pago:")
print(facturacion_por_pago)
````

#7: Facturaci√≥n por vendedor


````python
# Agrupar por vendedor y sumar la facturaci√≥n
facturacion_por_vendedor = tienda.groupby('Vendedor')['Precio'].sum().sort_values(ascending=False)

# Mostrar resultados
print("üßç‚Äç‚ôÇÔ∏è Facturaci√≥n por vendedor:")
print(facturacion_por_vendedor)
````

#8: Facturaci√≥n por ciudad o lugar de compra


````python
# Agrupar por lugar de compra y sumar la facturaci√≥n
facturacion_por_ciudad = tienda.groupby('Lugar de Compra')['Precio'].sum().sort_values(ascending=False)

# Mostrar resultados
print("üó∫Ô∏è Facturaci√≥n por ciudad:")
print(facturacion_por_ciudad)
````
